```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.rcParams['font.size'] = 30

# Generate data points for the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)  # mean=0, standard deviation=1

# Plot the standard normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='black', linewidth=2, label='Standard Normal Distribution')

# Define z-score intervals and corresponding shades
intervals = [(-3, -2), (-2, -1), (-1, 1), (1, 2), (2, 3)]
# shades = ['lightgrey', 'lightblue', 'lightgreen', 'lightcoral', 'lightpink']
shades = ['#9ecae1', '#6baed6', '#4292c6', '#6baed6', '#9ecae1']
# Shade regions between z-score intervals with varying shades
for interval, shade in zip(intervals, shades):
    x_interval = np.linspace(interval[0], interval[1], 1000)
    y_interval = norm.pdf(x_interval)
    plt.fill_between(x_interval, y_interval, color=shade)

# Label x-axis ticks with corresponding standard deviation values
plt.xticks(np.arange(-4, 5, 1), labels=['-4', '-3', '-2', '-1', '0', '1', '2', '3', '4'])

# Add labels and legend
plt.xlabel('z-score')
# plt.legend()
for line in [-3, -2, -1, 0, 1, 2, 3]:
    plt.plot([line, line], [0, norm.pdf(line)], color='black', linestyle='--', linewidth=1)
# plt.axvline(x=0, color='black', linestyle='--', linewidth=1)
# plt.axvline(x=1, color='black', linestyle='--', linewidth=1)
# plt.axvline(x=2, color='black', linestyle='--', linewidth=1)
# plt.axvline(x=3, color='black', linestyle='--', linewidth=1)
# plt.axvline(x=-1, color='black', linestyle='--', linewidth=1)
# plt.axvline(x=-2, color='black', linestyle='--', linewidth=1)
# plt.axvline(x=-3, color='black', linestyle='--', linewidth=1)
    
plt.annotate('', xy=(0, 0.1), xytext=(1, 0.1),
             arrowprops=dict(arrowstyle='<|-|>', color='black', lw=1.5),
             annotation_clip=False)
plt.text(0.5, 0.12, '1σ', horizontalalignment='center')

plt.tick_params(left=False, right=False, top=False, labelleft=False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.savefig('z_score_example.png', bbox_inches='tight', dpi=300)
```
